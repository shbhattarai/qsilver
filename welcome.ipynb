```python
some_qubit =[(1j-2)/2, -1/np.sqrt(2)]
print(some_qubit)
```

```python
sum =0
for el in some_qubit:
    sum = sum + np.abs(el)**2
print("Length of the vector: ",np.sqrt(sum))
```

```python
qubit2= [(2+1j)/4, np.sqrt(3)/4]
sum =0
for el in qubit2:
    sum = sum + np.abs(el)**2
print("Length of the vector: ",np.sqrt(sum))
```

```python
local_phase_angle = np.random.uniform(0,2*np.pi)
alpha = np.random.uniform(0,np.pi/2)

local_phase = np.e**(1j*local_phase_angle)
random_qubit = [np.cos(alpha),local_phase*np.sin(alpha)]

print(random_qubit)
```
